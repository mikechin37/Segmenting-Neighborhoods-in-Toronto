```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

#### Read wiki data into dataframe

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content)
table = soup.find_all('table')[0]
df = pd.read_html(str(table))[0]
df.head()
```

#### Clean up dataframe 

```python
df.Borough.replace("Not assigned", np.nan, inplace = True)
df.head()
```

```python
df['Neighbourhood'] = np.where(df['Neighbourhood'] == "Not assigned", df['Borough'], df['Neighbourhood'])
df.head()
```

```python
df.dropna(axis = 0, inplace = True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df2 = df.groupby('Postcode')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x)).rename("Hoods").reset_index()
df2.head(20)
#df3 = df2.to_frame()
#df3.reset_index()
#df3.head()
```

```python
df3 = df.merge(df2)
df3.head(20)
```

```python
df3.drop(['Neighbourhood'], errors = 'ignore')
df3.drop(df.columns[2], axis = 1, inplace=True)
df3.head()
df3.rename(columns = {'Hoods':'Neighbourhood'}, inplace = True)
df3.head()
```

```python
df3.drop_duplicates(inplace=True)
df3.head()
```

```python
csv_path = 'https://cocl.us/Geospatial_data'
df_lat = pd.read_csv(csv_path)
df_lat.head()
```

```python
df_lat.rename(columns = {"Postal Code":"Postcode"}, inplace=True)
df_lat.head()
```

```python
df_combined = df3.merge(df_lat, how = 'outer')
df_combined.head()
```

```python
df_toronto = df_combined[df_combined['Borough'].str.contains("Toronto")]
df_toronto.head()
```

#### Create a map of New York with neighborhoods superimposed on top.

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[43.653908, -79.384293], zoom_start=11)

# add markers to map
for lat, lng, label in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
